```python
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
```

```python
def f(x):
    return x
```

```python
interact(f,x=10,)
```

```python
def f2(x):
    return x**2
```

```python
interact(f,x=10)
```

```python
def f3(s):
    return s
```

```python
interact(f3,s=True)#with the boolean creates a check box
```

```python
def f4(a):
    return a
```

```python
interact(f4,a='Hello')
```

```python
@interact(x=True,y=1.0)
def g(x,y):
    return(x,y)
```

```python
@interact(x=True,y=fixed(1.0)) #fixed to stablish certain values enable to the user to manipulate
def g2(x,y):
    return(x,y)
```

```python
interact(f,x=widgets.IntSlider(min=-100,max=100,step=1,value=0))
```

```python
interact(f,x=(-100,100,1))#abreviation of  IntSlyter
```

```python
interact(f,x=(-10.0,10.0,.1))#to make them floats
```

```python
@interact(x=(0.0,20.0,0.5))
def h(x=5.0):
    return x
```

```python
interact(f,x=['hello','name','city'])#to create list as menu
```

```python
interact(f,x={'one':10,'two':20})#also a menu, but it make it with key as option an values as result
```
